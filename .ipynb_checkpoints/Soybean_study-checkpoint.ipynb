### Supply chain

- Deforestation and Land Use Change
- Soil Health
- Climate change
- Land Rights
- Water use and pollution
- Working conditions

The aim of the study is to try to dissaggreate the posible data insights to help determine the supply chain risks.

### What is soy?
It is an annual crop. As it is a plant that only produces one yield in each life cycle, this means that to increase yield and meet demand, more soy has to be planted, which requires more land. 
This circunstance is one of the limitants for the crop.

### Explore the data
According to the publication *IFRI, 2019, "Global Spatially-Disaggregated Crop Production Statistics Data for 2010 Version 2.0"* 
the SPAM2010 (Spatial Production Allocation Model) dataset has the purpose to disaggregate the crop statistics of agriculture production (harvested area, production, yield) within the different farming systems.

The data includes 42 crops and its statistics. With this, it is possible to know the importance of the soybean among other crops. 

The data has 5 variables:
- **Harvested area**  --> measured in ha. It represents the area where the crop is harvested.
- Physical Area
- Yield
- Production
- Value of production

The data to be exploded will be the harvested area of soybean crop according to the technology used.


#### Farming systems
Soybean yield vary according to the farming system used. The next graph explains the different farming system types.



The difference between them is the water supply. The irrigated system need higher water than the rainfed ones. 

Crop yields specifically change among farming system. In other words, depending on the technology, there is a greater impact on on soil nutrient loss, deforestation and water runoff.

The technologies are divided into different datasets. We will merge all in one filtering by AOI.
- _TI--> irrigated portion of crop
- _TH--> rainfed high inputs portion of crop
- _TL--> rainfed low inputs portion of crop
- _TS--> rainfed subsistence portion of crop

```python
df_TI = pd.read_csv('data/HarvestedArea_spam2010/spam2010V2r0_global_H_TI.csv', encoding='cp1252')
df_TL = pd.read_csv('data/HarvestedArea_spam2010/spam2010V2r0_global_H_TL.csv', encoding='cp1252')
df_TS = pd.read_csv('data/HarvestedArea_spam2010/spam2010V2r0_global_H_TS.csv', encoding='cp1252')
df_TH = pd.read_csv('data/HarvestedArea_spam2010/spam2010V2r0_global_H_TH.csv', encoding='cp1252')
```

To join the dataframes of the different harvested areas by farming system, we should concatenate and then apply a spatial filter to get only the values inside the AOI.
 - Due to memory problems, the dataframes will be created and filter lonely and then merged for the purpose
 - The null values of soybean will be out

------------

```python
#Irrigated
df_TI.head()
```

```python
#We need the x,y, tech_type, unit, soyb_i,name_cntr,name_adm1_name_adm2
df_soyb_irrigated = df_TI[['x','y','tech_type','unit','soyb_i','name_cntr','name_adm1','name_adm2']]
df_TI.count()

#Change column name of 'soyb_i' to 'soyb'
df_soyb_irrigated.columns=['soybean' if x=='soyb_i'else x for x in df_soyb_irrigated.columns]

#In column tech_type, change the values to the correct syntax
df_soyb_irrigated.loc[df_soyb_irrigated["tech_type"] == "I","tech_type"]="Irrigated farming system"

#Get only columns without null vallues
df_soyb_irrigated = df_soyb_irrigated[df_soyb_irrigated['soybean']!=0]

df_soyb_irrigated.head()
```

```python
#Rainfed-high
df_soyb_rainfedhigh = df_TH[['x','y','tech_type','unit','soyb_h','name_cntr','name_adm1','name_adm2']]
df_soyb_rainfedhigh.columns=['soybean' if x=='soyb_h'else x for x in df_soyb_rainfedhigh.columns]
df_soyb_rainfedhigh.loc[df_soyb_rainfedhigh["tech_type"] == "H","tech_type"]="Rainfed-high farming system"
df_soyb_rainfedhigh = df_soyb_rainfedhigh[df_soyb_rainfedhigh['soybean']!=0]

df_soyb_rainfedhigh.head()
```

```python
#Rainfed-low 
df_soyb_lowrainfed = df_TL[['x','y','tech_type','unit','soyb_l','name_cntr','name_adm1','name_adm2']]
df_soyb_lowrainfed.columns=['soybean' if x=='soyb_l'else x for x in df_soyb_lowrainfed.columns]
df_soyb_lowrainfed.loc[df_soyb_lowrainfed["tech_type"] == "L","tech_type"]="Rainfed-low farming system"
df_soyb_lowrainfed = df_soyb_lowrainfed[df_soyb_lowrainfed['soybean']!=0]

df_soyb_lowrainfed.head()
```

```python
#Rainfed susbsistence 
df_soyb_subsistence = df_TS[['x','y','tech_type','unit','soyb_s','name_cntr','name_adm1','name_adm2']]
df_soyb_subsistence.columns=['soybean' if x=='soyb_s'else x for x in df_soyb_subsistence.columns]
df_soyb_subsistence.loc[df_soyb_subsistence["tech_type"] == "S","tech_type"]="Subsistence farming system"
df_soyb_subsistence = df_soyb_subsistence[df_soyb_subsistence['soybean']!=0]

df_soyb_subsistence.head()
```

Once the dataframes are ready, we have to spatially filter to the AOI. To do that, first it is necessary yo give the spatial extension to the dataframes, converting them into geodataframes

```python
#the coordinates are in the columns x and y
gdf_irrigated = gpd.GeoDataFrame(df_soyb_irrigated, geometry=gpd.points_from_xy(df_soyb_irrigated.x, df_soyb_irrigated.y))

#to ensure that have the WGS84 coordinate system, set the CRS for all geodataframes
gdf_irrigated.set_crs(epsg=4326, inplace=True)

gdf_irrigated.head()
```

```python
#calculate the total values computed in the dataset for the world. These depend on the number of rows of the dataset. 
#This data is important in order to compare between the AOI and the world dataset
print("The total values are " + str(gdf_lowrainfed.shape[0]) + ' of soybean crop in the world')
```

```python
gdf_rainfedhigh = gpd.GeoDataFrame(df_soyb_rainfedhigh, geometry=gpd.points_from_xy(df_soyb_rainfedhigh.x, df_soyb_rainfedhigh.y))
gdf_rainfedhigh.set_crs(epsg=4326, inplace=True)

gdf_rainfedhigh.head()
```

```python
gdf_lowrainfed = gpd.GeoDataFrame(df_soyb_lowrainfed, geometry=gpd.points_from_xy(df_soyb_lowrainfed.x, df_soyb_lowrainfed.y))
gdf_lowrainfed.set_crs(epsg=4326, inplace=True)

gdf_lowrainfed.head()
```

```python
gdf_subsistence = gpd.GeoDataFrame(df_soyb_subsistence, geometry=gpd.points_from_xy(df_soyb_subsistence.x, df_soyb_subsistence.y))
gdf_subsistence.set_crs(epsg=4326, inplace=True)
gdf_subsistence.head()
```

Before getting all the data into one single geodataframe, we will apply the spatial filter by the bouding box defined by the AOI.
AOI is divided into two zones, so the spatial filter will be done in two steps and then merge.

```python
#Filter the data by the bounding box of the AOI
aoi.bounds
```

#### South Asia Region 

```python
#Bounds of South Asia region for the systems
sa_irrigated = gdf_irrigated.cx[70.664062:88.945312, 9.449062:30.448674]
sa_rainfedhigh = gdf_rainfedhigh.cx[70.664062:88.945312, 9.449062:30.448674]
sa_rainfedlow = gdf_lowrainfed.cx[70.664062:88.945312, 9.449062:30.448674]
sa_subsistence = gdf_subsistence.cx[70.664062:88.945312, 9.449062:30.448674]
```

#### Latin America Region 

```python
#Bounds of Latin America Region
la_irrigated = gdf_irrigated.cx[-68.203125:-49.570312, -28.921631:2.460181]
la_rainfedhigh = gdf_rainfedhigh.cx[-68.203125:-49.570312, -28.921631:2.460181]
la_rainfedlow = gdf_lowrainfed.cx[-68.203125:-49.570312, -28.921631:2.460181]
la_subsistence = gdf_subsistence.cx[-68.203125:-49.570312, -28.921631:2.460181]
```

```python
#Concatenate all the geodataframes
sa_tech = pd.concat([sa_irrigated,sa_rainfedhigh,sa_rainfedlow,sa_subsistence], ignore_index=True, sort=True)
la_tech = pd.concat([la_irrigated,la_rainfedhigh,la_rainfedlow,la_subsistence], ignore_index=True, sort=True)
```

```python
#Clip world boundaries with bbox chosen. This area is a bit bigger than the AOI to improve vizualisation
#Get bbox for outer part of AOI in Latin America and Caribbean Region
bbox_la = (-75.058594,-34.016242,-32.871094,7.623887)
polygon_la = shapely.geometry.box(*bbox_la, ccw=True)

#Get bbox for outer part of AOI in South Asia Region
bbox_sa = (69.565430,5.484768,93.911133,32.583849)
polygon_sa = shapely.geometry.box(*bbox_sa, ccw=True)

#Create geodataframe for each region
sa = gpd.GeoDataFrame(crs='epsg:4326', geometry= [polygon_sa])
la = gpd.GeoDataFrame(crs='epsg:4326', geometry= [polygon_la])

#Clip the world gdf to show the outputs
world_clipped_sa = world.clip(sa)
world_clipped_la = world.clip(la)

fig, ax = plt.subplots(figsize=(12, 8))
world_clipped_sa.plot(ax=ax, color="blue")
world_clipped_la.plot(ax=ax, color="blue")
world.boundary.plot(ax=ax)
aoi.boundary.plot(ax=ax, color="red")
ax.set_title("World boundaries with AOI", fontsize=20)
ax.set_axis_off()
plt.show()
```

```python
#Prepare plot for AOI
fig, (ax1,ax2) = plt.subplots(1,2)
world_clipped_la.boundary.plot(ax=ax1)
aoi_la= gpd.GeoSeries(aoi.iloc[1])
aoi_la.boundary.plot(ax=ax1, color="red")
ax1.set_title("Latin American and Caribbean Region", fontsize=10)
ax1.set_axis_off()
# ax1.plot(world_clipped_la,aoi_la)

world_clipped_sa.boundary.plot(ax=ax2)
aoi_sa= gpd.GeoSeries(aoi.iloc[0])
aoi_sa.boundary.plot(ax=ax2, color="red")
ax2.set_title("South Asia Region", fontsize=10)
ax2.set_axis_off()


plt.show()
```

```python
#Check if all farming technologies are in the two regions
print('In Latin America and Caribbean Region, there are the following farming systems ' + str(la_tech['tech_type'].unique()))
print('In South Asia , there are the following farming systems ' + str(sa_tech['tech_type'].unique()))
```

So, we know that in Latin America and Caribbean Region, there is only three farming systems
 - Irrigated farming system
 - Rainfed-high farming system
 - Rainfed-low farming system

Now, we will map and then calculate the surface of each one and compare them

In South Asia, all of the farming systems are used

 - Irrigated farming system
 - Rainfed-high farming system
 - Rainfed-low farming system
 - Subsistence farming system
 

```python
color_dict_la = {'Irrigated farming system':'aqua','Rainfed-high farming system':'orangered', 'Rainfed-low farming system':'forestgreen'}
color_dict_sa = {'Irrigated farming system':'aqua','Rainfed-high farming system':'orangered', 'Rainfed-low farming system':'forestgreen', 'Subsistence farming system':'yellow'}

fig, (ax_la,ax_sa) = plt.subplots(1,2, figsize=(12,8))
world_clipped_la.boundary.plot(ax=ax_la)
aoi_la= gpd.GeoSeries(aoi.iloc[1])
aoi_la.boundary.plot(ax=ax_la, color="red")
ax_la.set_title("Latin American and Caribbean AOI Region", fontsize=10)
la_tech.plot(ax=ax_la, categorical=True,column="tech_type", alpha=0.5, legend=True, marker='x', markersize= 0.05, cmap=colors.ListedColormap(list(color_dict_la.values())))
ax_la.set_axis_off()


world_clipped_sa.boundary.plot(ax=ax_sa)
aoi_sa= gpd.GeoSeries(aoi.iloc[0])
aoi_sa.boundary.plot(ax=ax_sa, color="red")
ax_sa.set_title("South Asia AOI Region", fontsize=10)
sa_tech.plot(ax=ax_sa, categorical=True,column="tech_type", alpha=0.5, legend=True, marker='x', markersize= 0.05, cmap=colors.ListedColormap(list(color_dict_sa.values())))
ax_sa.set_axis_off()
```

Apparently, the technology with higher surface is rainfed-high farming system. Let's check with the tables
